```python
import numpy as np
import matplotlib.pyplot as plt
mu = 0 # mean
sigma = 3  # standard deviation
```

```python

s = np.random.normal(mu, sigma, 100)  # Create Gaussian Noise

# Plot Gaussian Noise
plt.xlim(0, 100)
plt.ylim(-20, 20)
plt.scatter(np.linspace(1, 100, num=100), s)
```

```python
min(s)
```

```python
# Plot Gaussian Distribution
count, bins, ignored = plt.hist(s, 30, edgecolor='black')

# count: y-axis value
# bins : x-axis center value
# ignored: obj
```

```python
plt.plot(bins, 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')
```
