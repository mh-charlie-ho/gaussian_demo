```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(42)
X = np.random.randn(100)
Y = 2*X + 1 + 0.5*np.random.randn(100)

sns.scatterplot(x=X, y=Y)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X and Y')
```

```python
from scipy.stats import multivariate_normal

mean = [0, 0]
covariance_matrix = [[2, 0],
                     [0, 7]]
X, Y = np.random.multivariate_normal(mean, covariance_matrix, 1000).T

sns.kdeplot(x=X, y=Y, fill=True, levels=20)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Joint Probability Density Function (JPDF)')
```
